## Data Validation

Some of the columns seem to be computed, as they are not listed in the provided dictionary.png. As there is no documentation on why and how these have been computed, I want to investigate and validate these new columns.

### Z_CostContact and Z_Revenue

```python
print(df['Z_CostContact'].value_counts())
print(df['Z_Revenue'].value_counts())

# Both columns contain the same value for all observations.
# Matter of fact, the descriptive statitistics already gave that away, showing that these 2 columns have 0 variance and thus contain the same value in each row.


# -> CONCLUSION: Drop these variables, as they are constant and will not add to the insights in this project.
df.drop(['Z_CostContact', 'Z_Revenue'], axis=1, inplace=True)
```

### MntTotal and MntRegularProds

```python
# These two seem to be sums of the individual amounts spend on all the products.

Mnt_cols = [col for col in df.columns if col.startswith('Mnt')]
print(df.filter(Mnt_cols).head())
# On first sight, MntRegularProds seems to exclude the spendings from MntGoldProds.
# However, in the descriptive statistics, negative values were found for MntRegularProds.

# Create new columns including own calculations
new_Mnt_cols = ['MntTotal', 'MntRegularProds']
orig_Mnt_cols = df.filter(like='Mnt').drop(new_Mnt_cols, axis=1).columns.to_list()
reg_Mnt_cols = [x for x in orig_Mnt_cols if x != 'MntGoldProds']

df['myMntTotal'] = df.filter(orig_Mnt_cols).sum(axis=1)
df['myMntRegularProds'] = df.filter(reg_Mnt_cols).sum(axis=1)

df.filter(like='myMnt').describe()
# No negative values.

# Compare own calcs to existing col
print('No. of cols matching MntTotal:', sum(df['myMntTotal'] == df['MntTotal']))
print('No. of cols matching myMntRegularProds:', sum(df['myMntRegularProds'] == df['MntRegularProds']))
# Only 61 rows match for MntTotal and myMntRegularProds.

print('Same 61 rows with mathching cols?:', df.loc[df['myMntTotal'] == df['MntTotal']].equals(df.loc[df['myMntRegularProds'] == df['MntRegularProds']]))
# These 61 rows are the same for both checks

print('\nDESCRIPTIVE STATISTICS FOR CASES WITH MATCHING COLS\n', df.loc[df['myMntTotal'] == df['MntTotal']].filter(like='Mnt').describe())
# All of the matching cases have 0 spendings in gold.

# Check whether computed col MntTotal might be missing the MntGoldProds col
print('No. of cols matching MntTotal+MntGoldProds:', sum(df['myMntTotal'] == df['MntTotal'] + df['MntGoldProds']))
# All 2205 rows match!
print('No. of cols matching MntRegularProds+MntGoldProds:', sum(df['myMntRegularProds'] == df['MntRegularProds'] + df['MntGoldProds']))
# All 2205 rows match!


# -> CONCLUSION: Drop pre-computed values and keep only myMntTotal, i.e. the sum of the amont spent for all product categories. 
df.drop(['MntTotal', 'MntRegularProds', 'myMntRegularProds'], axis=1, inplace=True)
```

### AcceptedCmpOverall

```python
# This column presumably contains the number of all campaigns accepted, thus the sum of the 5 individiual AcceptedCmpX columns. 
df.filter(like='AcceptedCmp')

df['myAcceptedCmpOverall'] = df.filter(like='AcceptedCmp').drop('AcceptedCmpOverall', axis=1).sum(axis=1)

print('Percentage of rows matching AcceptedCmpOverall:', round(sum(df['myAcceptedCmpOverall'] == df['AcceptedCmpOverall']) / len(df) * 100, 0), '%')
# 100% match values in both columns

# CONCLUSION: Variable AcceptedCmpOverall is kept, my computed col is dropped. 
# NOTE: This column does not consider accepting the offer of the most recent campaign, which is stored in column 'Response'
df.drop('myAcceptedCmpOverall', axis=1, inplace=True)
```

## Variable computation

Additional variables will be computed:
1. no_of_kids = Number of Kids (sum of Kidhome + Teenhome)
2. has_kids = Kids Yes/No (1 if no_of_kids > 0, else 0)
3. NumPurchasesOverall = Sum of NumDealsPurchases, NumCatalogPurchases, NumStorePurchases, NumWebPurchases

### 1.+2. no_of_kids and has_kids

```python
df['no_of_kids'] = df['Kidhome'] + df['Teenhome']
df['has_kids'] = np.where(df['no_of_kids'] > 0, 1, 0)

print(pd.crosstab(df['Kidhome'], df['Teenhome']))
print(pd.crosstab(df['no_of_kids'], df['has_kids']))

df.filter(['Kidhome', 'Teenhome', 'no_of_kids', 'has_kids']).describe()
```

### NumPurchasesOverall

```python
num_purch_cols = df.filter(like='Purchase').columns.to_list()

df['NumPurchasesOverall'] = df.filter(num_purch_cols).sum(axis=1)

print(df.filter(like='Purchase').head())
df.filter(like='Purchase').describe()
```

### Accepted Campaigns

```python
Acc_cols = df.filter(like='Acc').columns.to_list()
Acc_cols.remove('AcceptedCmpOverall')
Acc_cols.append('Response')
Acc_cols.sort()

# Plot
fig, axs = plt.subplots(figsize=(18, 16), ncols=len(Acc_cols)-1, nrows=len(Acc_cols)-1)
#plt.subplots_adjust(wspace=0.5)

for ax_y, y in enumerate(Acc_cols):
    for ax_x, x in enumerate(Acc_cols):
        if ax_y == ax_x:
            for del_x in range(ax_x+1, len(Acc_cols)-1):
                fig.delaxes(axs[ax_y, del_x])
            break
        cur_ax = axs[ax_y-1, ax_x]
        ct = pd.crosstab(df[y], df[x])
        jc = round(ct[1][1] / (ct[0][1]+ct[1][0]+ct[1][1]), 2)
        sns.heatmap(ct, square=True, annot=True, fmt='d', ax=cur_ax, linewidths=0.5, cmap=sns.diverging_palette(20, 220, n=200))
        cur_ax.text(1, 0.95, f'Jaccard Similarity:', horizontalalignment='center')
        cur_ax.text(1, 1.15, f'{jc:.2f}', weight='bold', horizontalalignment='center')
```

### Variables Split by Response

#### Dichotomous Variables

```python
all_cols = df.drop('Response', axis=1).columns.to_list()

# Extract columns names of variables with binary responses
bin_cols = df.filter(like='Accepted').columns.to_list()
bin_cols.remove('AcceptedCmpOverall')
bin_cols.extend(['has_kids', 'Complain'])
bin_cols.extend(df.filter(like='education').columns.to_list())
bin_cols.extend(df.filter(like='marital').columns.to_list())

# Plot
fig, axs = plt.subplots(figsize=(18, 18), ncols=6, nrows=3)
plt.subplots_adjust(wspace=0.5)
axs = axs.ravel()

for axis, col in enumerate(bin_cols):
    sns.countplot(x=df[col], hue=df['Response'], ax=axs[axis])
    ct = pd.crosstab(df[col], df['Response'])
    share1 = round((ct[1][0] / (ct[0][0]+ct[1][0]))*100, 1)
    share2 = round((ct[1][1] / (ct[0][1]+ct[1][1]))*100, 1)
    y_max = max(ct[0][0], ct[0][1], ct[1][0], ct[1][0])
    axs[axis].text(0, y_max/4, f'{share1}%', weight='bold', horizontalalignment='center')
    axs[axis].text(1, y_max/4, f'{share2}%', weight='bold', horizontalalignment='center')
fig.delaxes(axs[len(axs)-1])
```

#### Numerical Variables

```python
# Collect rest (i.e. the numeric cols) in one list
cols = [col for col in all_cols if col not in bin_cols]

# Plot
fig, axs = plt.subplots(figsize=(18, 18), ncols=7, nrows=3)
plt.subplots_adjust(wspace=0.5)
axs = axs.ravel()

for axis, col in enumerate(cols):
    sns.boxplot(x=df['Response'], y=df[col], ax=axs[axis])
```
